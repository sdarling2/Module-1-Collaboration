```python
## Testing Your Code

Testing is the process of checking whether code works as expected.

There are two main types of testing:

- **Manual testing**: Running the program and checking results yourself.
- **Automated testing**: Using code to test code.

Automated testing is more reliable because it can be repeated quickly and consistently.

### Unit Testing vs Integration Testing

- **Unit testing** checks a single function or component in isolation.
- **Integration testing** checks how multiple components work together.

Unit tests help identify exactly where errors occur, making debugging easier.
```

```python
import unittest

class TestSum(unittest.TestCase):

    def test_sum(self):
        self.assertEqual(sum([1, 2, 3]), 6)

    def test_sum_tuple(self):
        self.assertEqual(sum((1, 2, 2)), 6)

print("Tests written (run in terminal using unittest)")
```

```python
## Writing Your First Test

A unit test is created by:

1. Defining a test class that inherits from `unittest.TestCase`
2. Writing test methods that start with `test_`
3. Using assertion methods like `assertEqual` to compare expected and actual results
```

```python
## Executing Your First Test

The tests were executed using the terminal command:

python test_sum_unittest.py

The unittest framework discovered and ran the test methods automatically and displayed whether each test passed or failed.
The test results show that two tests were executed. One test passed and one test failed.
The passing test confirms that the function correctly calculates the sum of `[1, 2, 3]` as 6. The failing test occurs because the expected result was incorrect: the actual sum of `(1, 2, 2)` is 5, not 6.
The output includes helpful debugging information such as the test name, the line where the failure occurred, and the expected vs actual values. This makes it easier to identify and fix errors in the code.
```

```python
## Running Tests With nose2

nose2 is a test runner that automatically discovers and runs unittest-based tests.
It works by scanning files named `test*.py` and executing any test classes that inherit from `unittest.TestCase`.
To install and run nose2:
python -m pip install nose2
python -m nose2
```

```python
## Running Tests With pytest

pytest is a more modern and simpler testing framework compared to unittest.
It allows tests to be written as simple functions using Python’s built-in `assert` statement instead of requiring classes.
Example:

def test_sum():
assert sum([1, 2, 3]) == 6

def test_sum_tuple():
assert sum((1, 2, 2)) == 6

To run pytest:
python -m pytest


pytest automatically discovers test files and provides detailed failure reports.
```

```python
## Findings

Unit testing helps ensure that individual parts of a program work correctly. Using frameworks like unittest, nose2, and pytest allows developers to automate testing and quickly identify errors.

Each framework provides different levels of simplicity and functionality, with pytest being the most modern and easiest to use.
```
