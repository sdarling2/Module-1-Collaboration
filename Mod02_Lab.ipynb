```python
# Name: Stephanie Scott
# File Name: Mod02_Lab.py
# Description: This program accepts student first name/last name and corresponding GPA, 
#              then determines whether the student qualifies for the Dean's List or Honor Roll.

while True:
    last_name = input("Enter student's last name (or 'ZZZ' to quit): ")
    
    if last_name == 'ZZZ':
        break

    first_name = input("Enter student's first name: ")
    gpa = float(input("Enter student's GPA: "))

    if gpa >= 3.5:
        print(first_name, last_name, "has made the Dean's List.")
    elif gpa >= 3.25:
        print(first_name, last_name, "has made the Honor Roll.")
    else:
        print(first_name, last_name, "does not qualify for the Dean's List or Honor Roll.")

    print()  # blank line for readability
```
